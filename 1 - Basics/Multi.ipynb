```python
from shapely.geometry import Point, LineString, Polygon, MultiPoint, MultiLineString, MultiPolygon
```

### Collection of multiple objects

#### Multiple points

```python
p1 = Point(0, 0)
p2 = Point(2, 4)
p3 = Point(4, 2)
p4 = Point(6, 6)
```

```python
ps = MultiPoint([p1, p2, p3, p4])
ps
```

#### Multiple lines

```python
l1 = LineString([p1, p2, p3, p4])
l2 = LineString([(6, 6), (8, 2), (10, 4), (12, 0)])

l1
```

```python
l2
```

```python
ls = MultiLineString([l1, l2])
ls
```

#### Multiple polygons

```python
c1 = Point((-5, 0)).buffer(4)
c2 = Point((5, 0)).buffer(4)
l3 = Polygon([(-1, 0.5), (-1, -0.5), (1, -0.5), (1, 0.5)])

gla = MultiPolygon([c1, l3, c2])
gla
```

⭐ Smallest rectangle that can fully surround it

```python
gla.envelope
```

⭐ [Convex hull](https://en.wikipedia.org/wiki/Convex_hull)

```python
gla.convex_hull
```
