```python
from shapely.geometry import Polygon, LinearRing, Point
```

### Creating polygons!

#### Normal polygons

⭐ Triangle

```python
trig = Polygon([(0, 0), (3, 0), (0, 4)])
trig
```

⭐ Rectangle

```python
rect = Polygon([(-2, 0), (2, 0), (2, 2), (-2, 2)])
# Try this instead and see what happens!
# Polygon([(0, 0), (2, 0), (0, 2), (2, 2)])
rect
```

⭐ Others

```python
val = 3**.5    # sqrt(3)
poly_1 = Polygon([(0, 0), (1, 0), (1.5, val/2), (1, val), (0, val), (-0.5, val/2)])
poly_1    # my cute little regular hexagon
```

#### Polygon with a hole!

```python
print(rect.exterior)
print(list(rect.interiors))
```

```python
poly_2 = Polygon(rect.exterior, [poly_1.exterior])
poly_2
```

#### Circle

```python
circle = Point((0.5, val/2)).buffer(1)
circle
```

```python
poly_3 = Polygon(circle.exterior, [poly_1.exterior])
poly_3
```
